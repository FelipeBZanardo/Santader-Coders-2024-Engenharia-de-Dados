Definimos uma função para essa operação.

```python
def slug_into_dataframe(owid_slug_value = '', output_file_folder = '', verbose = False, output_file = False):

    """
    Uma função para baixar dados presentes em gráficos (`charts`)
    """
    
    owid_link_slug_value = f'https://ourworldindata.org/grapher/{owid_slug_value}'        

    #pdf_owid = charts.get_data(owid_slug_value)
    pdf_owid = charts.get_data(owid_link_slug_value)

    if verbose == True:
        #print('type(df):', type(pdf_owid))
        print('owid_link_slug_value:', owid_link_slug_value)
        #print('pdf_owid:', owid_slug_value)
        #display(pdf_owid)

    if output_file == True:
        output_file_name = os.path.join(output_file_folder, owid_slug_value)
        print('output_file_name:' , output_file_name)
        pdf_owid.to_pickle(output_file_name + ".pkl")
        
    return pdf_owid
```

```python
def slugs_intersect_columns(dict_slug_column_names):

    """
    Função para encontrar as columnas em comum em um conjunto de dataframes
    """
    return list(reduce(lambda s1, s2: s1.intersection(s2), list(dict_slug_column_names.values())))
```

```python
def slugs_merge_dataframes(
        pdf_slug_merge,
        list_search_value = [],
        verbose = True,
        ):
    for title in list_search_value:

        #print('title:', title)

        owid_slug_value = title

        pdf_slug = slug_into_dataframe(owid_slug_value = owid_slug_value, verbose = False, output_file = False) 

        pdf_slug_merge = pd.merge(
            left = pdf_slug_merge, 
            right = pdf_slug, 
            left_on = slugs_intersect_columns, 
            right_on = slugs_intersect_columns,
            how = 'outer',
            #left_index = True,
            #right_index = True, 
            suffixes = ('', '_remove')
            )
        pdf_slug_merged_sorted = pdf_slug_merge.sort_values(by = slugs_intersect_columns, ascending = [True, True], inplace = False)
        
        if verbose == True:

            print('pdf_slug:', title)
            display(pdf_slug.head())
            display(pdf_slug.tail())
            #print('pdf_slug_merge', pdf_slug_merge)
            #display(pdf_slug_merge.head())
            #display(pdf_slug_merge.tail())

    for column in list(pdf_slug_merge.columns):

        if verbose == True: print('column:', column)

        if '_remove' in column: 

            if verbose == True: print(f'\t remove column {column}') 

            del pdf_slug_merge[column]

        if verbose == True: print('')
        
    if verbose == True:
        print('pdf_slug_merged_sorted.shape:', pdf_slug_merged_sorted.shape) 
        print('pdf_slug_merged_sorted.info():')
        display(pdf_slug_merged_sorted.info())
        print('pdf_slug_merged_sorted.describe():')
        display(pdf_slug_merged_sorted.describe())        
        print('pdf_slug_merged_sorted:')
        display(pdf_slug_merged_sorted.head(10))
        display(pdf_slug_merged_sorted.tail(10))

    return pdf_slug_merge, pdf_slug_merged_sorted
```

```python
def check_tables_in_database(
        db_name,        
        ):

    query = '''
    SELECT 
        name
    FROM 
        sqlite_schema
    WHERE 
        type ='table' AND 
        name NOT LIKE 'sqlite_%';
    '''

    connection = sqlite3.connect(db_name)

    print('connection.total_changes:', connection.total_changes)

    cursor = connection.cursor()
    #cursor.execute(query)
    tables = cursor.execute(query).fetchall()    
    cursor.close()
    
    print(f'tables in database {db_name}:', tables)

    return tables
```

```python
def read_table_in_database(
        db_name,
        table_name = '',
        n_rows = 20
        ):

    query = f'''
    SELECT 
        * 
    FROM 
        {table_name}
    '''

    connection = sqlite3.connect(db_name)
    cursor = connection.cursor()
    table = cursor.execute(query).fetchmany(n_rows)
    cursor.close()

    display(table)
```

```python
import os

import pandas as pd
import pandas_access as mdb

import numpy as np

import requests

import json

from matplotlib import pyplot as plt

import seaborn as sns

# owid
from owid.catalog import charts
from owid import catalog

from functools import reduce
```

```python
class OWID:

    def __init__(self, owid_slug_value, search_value):

        self.owid_slug_value = owid_slug_value
        self.search_value = search_value

    def slug_into_dataframe(self, verbose = False, output_file = False):
        """
        Uma função para baixar dados presentes em gráficos (charts)
        """

        output_file_folder = 'OWID_Data'

        slug = self.owid_slug_value
        owid_link_slug_value = f'https://ourworldindata.org/grapher/{slug}'
        pdf_owid = charts.get_data(slug)

        if verbose == True: print('owid_link_slug_value:', owid_link_slug_value)
        if output_file == True:
            output_file_name = os.path.join(output_file_folder, self.owid_slug_value)
            print('output_file_name:', output_file_name)
            pdf_owid.to_pickle(output_file_name + ".pkl")
        
        return pdf_owid
    
    def slugs_search_value_in_list(
            self,
            verbose = False,
            ):

        list_slug_topics = charts.list_charts()
        list_search_output = []

        for title in list_slug_topics:
            #print('title:', title)
            
            for world in title.split('-'):
                #print('world:', world)

                if world == self.search_value: list_search_output.append(title)

        if verbose == True:
            print('self.search_value:', self.search_value)
            print('')
            print("len(list_search_output):", len(list_search_output)) 
            print('list_search_output:')
            display(list_search_output)
            
        return list_search_output    

    @classmethod        
    def slugs_columns_to_dictionary(
            cls,
            list_search_value,
            verbose = False,
            ):
        
        set_common_columns = set()
        dict_slug_column_names = {}
        all_columns = []
        for title in range(0, len(list_search_value), 1)[0 : 3]:
            owid_slug_value = list_search_value[title]

            slug = owid_slug_value
            pdf_slug = charts.get_data(slug)

            dict_slug_column_names.update({list_search_value[title] : set(pdf_slug.columns)})
            all_columns.append(list(pdf_slug.columns))
            if verbose == True:
                #print('title:', title)
                print(f'list_search_value[{title}]', list_search_value[title])
                display(pdf_slug.head())
                display(pdf_slug.tail())
        if verbose == True:
            print('dict_slug_column_names:')
            display(dict_slug_column_names)
            print('all_columns:')
            display(all_columns)
            #unique_columns = pd.Series(all_columns).unique()

        return dict_slug_column_names, all_columns
    
    def slugs_intersect_columns(cls, dict_slug_column_names):

        """
        Função para encontrar as columnas em comum em um conjunto de dataframes
        """

        return list(reduce(lambda s1, s2: s1.intersection(s2), list(dict_slug_column_names.values())))

    def slugs_difference_columns(cls, list_of_lists):
    
        return list(reduce(lambda x, y: x + y, list_of_lists, []))
    
    def slugs_merge_dataframes(
            cls,
            pdf_slug_merge,
            list_search_value = [],
            slugs_intersect_columns = [],
            verbose = False,
            ):
        
        '''Método para fundir dataframes com tópicos semelhantes'''

        for title in list_search_value:

            owid_slug_value = title
            slug = owid_slug_value
            pdf_slug = charts.get_data(slug)

            pdf_slug_merge = pd.merge(
                left = pdf_slug_merge,
                right = pdf_slug,
                left_on = slugs_intersect_columns,
                right_on = slugs_intersect_columns,
                how = 'outer',
                suffixes = ('', '_remove'),
                )
            

            if verbose == True:

                print('pdf_slug:', title)
                display(pdf_slug.head())
                display(pdf_slug.tail())
                #print('pdf_slug_merge', pdf_slug_merge)
                #display(pdf_slug_merge.head())
                #display(pdf_slug_merge.tail())
            
        for column in list(pdf_slug_merge.columns):

            if '_remove' in column:

                if verbose == True: print(f'\t remove column {column}')

                del pdf_slug_merge[column]
                del pdf_slug_merged_sorted[column]

            if verbose == True: print('')
        
        pdf_slug_merged_sorted = pdf_slug_merge.sort_values(by = slugs_intersect_columns, ascending = True, inplace = False)

        if verbose == True:
            print('pdf_slug_merged_sorted.shape:', pdf_slug_merged_sorted.shape) 
            print('pdf_slug_merged_sorted.info():')
            display(pdf_slug_merged_sorted.info())
            print('pdf_slug_merged_sorted.describe():')
            display(pdf_slug_merged_sorted.describe())        
            print('pdf_slug_merged_sorted:')
            display(pdf_slug_merged_sorted.head(10))
            display(pdf_slug_merged_sorted.tail(10))

        return pdf_slug_merge, pdf_slug_merged_sorted

    def find_search_value_in_pdf(
            cls,
            search_value,
            verbose = False,
            ):

        owidccCF_find_search = catalog.find(search_value)

        if verbose == True:
            print('type(owidccCF_find_search):', type(owidccCF_find_search))
            print('owidccCF_find_search')
            display(owidccCF_find_search.head())
            display(owidccCF_find_search.tail())

        pdf_find_search = pd.DataFrame(owidccCF_find_search)

        dict_find_search_dimensions = dict()

        for dimension_list, table_name in zip(pdf_find_search['dimensions'], pdf_find_search['table']):
            
            dict_find_search_dimensions.update({table_name : set(dimension_list)})

        return pdf_find_search, dict_find_search_dimensions
    
    def find_intersect_columns(cls, dict_find_search_dimensions):

        """
        Função para encontrar as columnas em comum em um conjunto de dataframes
        """
        return list(reduce(lambda s1, s2: s1.intersection(s2), list(dict_find_search_dimensions.values())))
    
    @classmethod
    def find_merge_by_common_columns(
        cls,
        list_find_paths,
        pdf_find_merge,
        find_intersect_columns = [],
        verbose = False,
        ):


        for path in list_find_paths:

            if verbose == True: print ('path:', path)

            rc = catalog.RemoteCatalog()
            table_find_search = rc[path]
            pdf_find_search_dataframe = pd.DataFrame(table_find_search)
            pdf_find_search_dataframe = pdf_find_search_dataframe.reset_index(inplace = False, drop = False)

            pdf_find_merge = pd.merge(
                left = pdf_find_merge, 
                right = pdf_find_search_dataframe, 
                left_on = find_intersect_columns, 
                right_on = find_intersect_columns,
                how = 'outer',
                #left_index = True,
                #right_index = True, 
                suffixes = ('', '_remove')
                )

        for column in list(pdf_find_merge.columns):

            if verbose == True: print('column:', column)

            if '_remove' in column: 

                if verbose == True: print(f'\t remove column {column}') 

                del pdf_find_merge[column]

        return pdf_find_merge                        
```

```python
import sqlalchemy
print('sqlalchemy.__version__:', sqlalchemy.__version__)
import sqlite3 
from sqlalchemy import create_engine
import os
```

```python
class DataBaseSQLite3:

    def __init__(self, db_name):
       
       self.db_name = db_name

    def create_database(self):
        ''' Método para criar uma base de dados local em SQLite3'''

        try: 
          conn = sqlite3.connect(self.db_name) 
          print(f"Database {self.db_name} formed.") 
          cursor = conn.cursor()
          print(dir(cursor))
        except sqlite3.Error as error: 
          print(f"Database {self.db_name} not formed.")
          print(f"Failed to connect with {self.db_name} database", error)
          
    def send_dataframe_to_database(
                    self,
                    dataframe,
                    table_name = '',
                    ):

            db = self.db_name
            db_engine = create_engine(f'sqlite:///{db}')

            # Store DataFrame in SQLite        
            dataframe.to_sql(
                name = table_name, 
                con = db_engine, 
                if_exists = 'replace', 
                index = False
                )

            connection = sqlite3.connect(self.db_name)        

            print(connection.total_changes)

    def check_tables_in_database(self):

        query = '''
        SELECT 
            name
        FROM 
            sqlite_schema
        WHERE 
            type ='table' AND 
            name NOT LIKE 'sqlite_%';
        '''

        connection = sqlite3.connect(self.db_name)
        print('connection.total_changes:', connection.total_changes)

        conn = sqlite3.connect(self.db_name)
        cursor = conn.cursor()
        #cursor.execute(query)
        tables = cursor.execute(query).fetchall()    
        #cursor.close()

        print(f'tables in database {self.db_name}:', tables)            

    def read_table_in_database(
            self,
            table_name = '',
            n_rows = 20
            ):

        query = f'''
        SELECT 
            * 
        FROM 
            {table_name}
        '''

        connection = sqlite3.connect(self.db_name)
        cursor = connection.cursor()
        table = cursor.execute(query).fetchmany(n_rows)
        cursor.close()
        display(table)

    def delete_database(self):

        if os.path.exists(self.db_name):
            os.remove(self.db_name)
            print("Successfully! The File has been removed")
        else:
            print("Can not delete the file as it doesn't exists")
            #pass
        






#    def execute_query_in_database_table(
#            self,
#            query = 
#            n_rows = 20
#            ):
#
#        connection = sqlite3.connect(self.db_name)
#        cursor = connection.cursor()
#        table = cursor.execute(query).fetchmany(n_rows)
#        cursor.close()
#        display(table)    
```
