## Comandos:

2. Acessar:

https://gitforwindows.org/

1. Baixar e instalar o Git.

2. Comandos do GIT:

Comandos Iniciais:

    - git config --global user.name "Lucio Monteiro"

    - git config --global user.email ljsmonteiro@gmail.com

    - git config --list

Criando repositorio vazio:

    - Escolher uma pasta e crie um arquivo qualquer.

    - git init (cria uma pasta .git)

    - git status

Adicionando arquivos ao repositório

    - git add .

    - git status

    - git commit -m "Arquivo inicial de citacoes"

    - git status

Alteração no aquivo

    - Faça alteração no arquivo

    - git status

    - git add .

    - git commit -m "Inserindo informações no aquivo - 1"

    - git log

    - git log --graph

    - git shortlog

    - git shortlog -sn

    - git show <hash>


Git diff

    - Editar um arquivo:

        Exemplo: vim README.md

    - git diff

    - git diff --name-only

    - git commit -am "Edit Readme"

    - git show

Desfazendo Edições

    1. Sem Add o arquivo

    - Editar um arquivo:

        Exemplo: vim README.md

    - git status

    - git checkout README.md

    - git diff

    2. Add o arquivo

    - Editar um arquivo:

        Exemplo: vim README.md

    - git add README.md

    - git diff

    - git reset HEAD README.md

    - git diff

    - git checkout README.md

    - git status

    3. Commit o arquivo

     - Editar um arquivo:

        Exemplo: vim README.md

    - git status

    - git commit -am "sjhskdhskh"

    - git log

    - git reset --soft <hash que anterior ou que vc quer ir> pronto para ser comitado

    - git status

    - git commit -am "kkkkkkk"

    - git log

    - git reset --mixed <hash que anterior ou que vc quer ir> precisa ser ADD

    - git diff

    - git commit -am "Trash"

    - git log

    - git reset --hard <hash que anterior ou que vc quer ir> pronto para ser comitado
 
    - git log

    - git status

3. GITHUB

- Criar um novo repositório

- Copiar as instruções do Repositório:

- Exemplo: 

    - git remote add origin https://github.com/ljsmonteiro/aula-git2.git
    - git branch -M main
    - git push -u origin main

- git remote

- git remote -v

4. Obtendo projeto Github

- Criar um novo diretório

- git clone https://github.com/ljsmonteiro/aula-git
