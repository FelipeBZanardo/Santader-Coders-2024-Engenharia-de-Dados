### Tópicos de estudo:

> ### 1. **Data Clean, Data Quality e Testes de Qualidade**;
> Aula 1: Por que precisamos de camadas harmonizadas e especializadas? 
> - Conceitos:
>> - Apresentação dos principais papéis dos especialistas de dados ao longo da pipeline de tratamento de dados (modelagem, engenharia de dados, analytics engineer, cientista de dados, analista de dados, engenheiro de ml). 
>> - Arquitetura de soluções de especialização de dados - Conceitos de armazenamento de dados com as camadas `raw` (dados brutos), `trusted` (dados harmonizados e/ou estruturados) e `spec` (dados especializados)
>> - Aplicações práticas de tabelas e visões especializadas e seus benefícios, tais como dashboards mais rápidos e com menor custo de processamento. 
> - Metodologia, Expositiva com os conceitos listados:
>> - Apresentação inicial das técnicas e ferramentas que serão ensinadas no módulo, com destaque que as camadas de armazenamento propostas neste material podem ser encontradas com diferentes nomes, tais como bronze, prata, ouro, raw, struct, refined, entre outras.
>> - Apresentar casos de uso envolvendo dashboards lentos e sobrecarregados prejudicando a tomada de decisão no tempo adequado e como camadas `trusted` e `spec` são essenciais em uma boa arquitetura de dados.
>> - Lembrar do tempo necessário para: Apresentação do módulo e validação de conhecimento mínimo. 
>> - Tarefa para casa (ou prático em sala): Instalação e configuração do ambiente com jupyter notebook funcional e servidor PostgreSQL, bem como pesquisa de conjuntos de dados (datasets) que possam servir de casos práticos de uma camada `raw` nos exercícios futuros. Como sugestão, no material de apoio, é apresentado um projeto de treino envolvendo extração via API até disponibilização em camada especializada.

> ### 2. **Geração de camadas especializadas**;
> Aula 2: Como um dado sujo atrapalha uma boa decisão?
> - Conceitos:
>> - Modelagem de tabelas especializadas com dados agrupados e/ou sumarizados.
>> - Princípios de qualidade de dados e seus benefícios.
>> - Técnicas de limpeza de dados para geração de tabelas harmonizadas e especializadas. 
> - Metodologia, Expositiva com os conceitos listados:
>> - Gerar debates a respeito de qualidade de dados apresentando casos práticos envolvendo dados divergentes.
>> - Exercícios (individual ou grupo): Validação de instalação do ambiente PostgreSQL para que seja possível seguir com a proposta: Modelar tabelas harmonizadas e especializadas com base nos conjuntos de dados apresentados como camada `raw` solicitados na aula anterior.

> Aula 3: Como garantir a completude do dado?
> - Conceitos:
>> - Técnicas de validação de integridade e qualidade de dados, tais como tratamento de duplicidade, nulidade, ausência de registros ou colunas, bem como divergência de tipagem
> - Metodologia, Expositiva com os conceitos listados:
>> - Expositiva com os conceitos listados, apresentando casos práticos em que dados duplicados ou com tipagem errada prejudicam a tomada de decisão.
>> - Exercícios (individual ou grupo): Em sequência à aula anterior, receber os resultados das modelagens e identificar as necessidades de tratamento de limpeza dos dados para persistência na camada harmonizada.

> ### 3. **Great Expectations**;
> Aula 4: Os dados atendem às expectativas?
> - Conceitos:
>> - Framework Great Expectations em Python.
>> - Técnicas de customização das regras de validação, documentação e perfil dos dados através do Great Expectations.
> - Metodologia, Expositiva com os conceitos listados:
>> - Expositiva com live coding: Apresentação do GE com implantação e uso, bem como cenários de testes validando as expectativas e implementação durante a execução da pipeline.
>> - Trabalho em grupo: Exercícios de validação utilizando o frame GE.
> ### 4. **Data Build Tool - dbt**.
> Aula 5: Como deixar de validar manualmente?
> - Conceitos:
>> - Apresentação, instalação e configuração do dbt-core.
>> - Uso do dbt para preparação de dados e construção das tabelas e visões especializadas.
>> - Configuração arquivos e pastas iniciais para funcionamento do dbt-core.
>> - Implementação de modelos básicos das tabelas via dbt-core.
> - Metodologia, Expositiva com os conceitos listados:
>> - Expositiva com os conceitos listados, apresentando casos práticos em que dados duplicados ou com tipagem errada prejudicam a tomada de decisão.
>> - Exercícios (individual ou grupo): Instalar e conectar o dbt-core à instância PostgreSQL instalada inicialmente para aplicar a construção das tabelas modeladas.
>> - Apresentação do desafio do projeto final e formação dos grupos.

> Aula 6: Como preparar pequenos tijolos fundamentais para validação de dados?
> - Conceitos:
>> - Técnicas de validação de dados via dbt, como renomeação, casting, sumarizações e categorizações.
>> - Loops e variáveis através do Jinja no dbt
> - Metodologia, Expositiva com os conceitos listados:
>> - Expositiva com os conceitos listados.
>> - Exercícios (individual ou grupo): Criar modelos especializados com base nos exercícios anteriores aplicando o uso das técnicas aprendidas de validação e Jinja.

> Aula 7: Como deixar a área de negócios mais satisfeita?
> - Conceitos:
>> - Uso do dbt para preparação de dados e construção das tabelas e visões especializadas.
>> - Criação de modelos materializados para geração das tabelas agrupamentos, sumarizações, categorizações ou outras especializações.
> - Metodologia, Expositiva com os conceitos listados:
>> - Expositiva com live coding.
>> - Trabalho em grupo: tempo em conjunto com professores para evolução do projeto final.

> Aula 8: Relembrando tudo
> - Conceitos:
>> - Revisão dos tópicos do módulo
> - Metodologia, Expositiva com os conceitos listados:
>> - Primeira hora exposição do resumo do que foi visto no módulo;
>> - Debate com tira-dúvidas (ao vivo ou por meio de um formulário/enquete)
>> - Trabalho em grupo: tempo em conjunto com professores para evolução do projeto final.

> Aula 9: Aula dedicada à devolutiva da avaliação por rubrica / autoavaliação
>> - Revisar os projetos de cada grupo salientando os pontos fortes e orientando oportunidades de melhoria
