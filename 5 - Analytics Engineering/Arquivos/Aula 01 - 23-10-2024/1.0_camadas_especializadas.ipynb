### Padroniza o nome das colunas da tabela bronze

```python
df_silver.columns = df_silver.columns.str.replace(r'\W', '_', regex=True)
df_silver.columns = df_silver.columns.str.replace("__", "_")
df_silver
```

### Adiciona uma nova coluna calculada e efetua uma limpeza simples

```python
df_silver["diff_high_low"] = df_silver["2_high"] - df_silver["3_low"]
df_silver
```

```python
df_silver = df_silver.dropna(subset=["datetime"])
df_silver
```

```python
values = {"1_open": 0, "2_high": 0, "3_low": 0, "4_close": 0, "5_volume": -1}
df_silver.fillna(value=values)
df_silver
```

```python
df_silver.dtypes
```

### Define os tipos de dados e armazena os resultados como tabela silver

```python
import sqlalchemy as sqlal
dict_dtype_silver={'datetime': sqlal.DateTime(), 
             '1_open':  sqlal.types.Float(precision=5, asdecimal=True),
             '2_high': sqlal.types.Float(precision=5, asdecimal=True),
             '3_low': sqlal.types.Float(precision=5, asdecimal=True),
             '4_close': sqlal.types.Float(precision=5, asdecimal=True),
             '5_volume': sqlal.types.INTEGER(),
             'diff_high_low': sqlal.types.Float(precision=5, asdecimal=True)}
```

```python
df_silver.to_sql("ibm_prices_silver_lauro", engine, if_exists="replace", index=False, dtype= dict_dtype_silver)
```

### Leitura da tabela silver

```python
query = """
SELECT * 
FROM ibm_prices_silver_lauro
"""
df_gold = pd.read_sql(sql=sql_text(query), con=engine.connect())
df_gold
```

```python
df_gold.dtypes
```

### Calcula os resultados por dia

```python
df_gold["date"] = df_gold["datetime"].dt.date
```

```python
df_gold = df_gold.groupby("date").agg(max_high=("2_high", "max"),
                                      min_low=("3_low", "min"),
                                      mean_diff_high_low=("diff_high_low", "mean"))
df_gold
```

```python
df_gold = df_gold.reset_index().rename({"index":"date"})
df_gold
```

```python
df_gold.dtypes
```

### Prepara os tipos de cada coluna e armazena os dados na camada Gold

```python
dict_dtype_gold = {'date': sqlal.Date(), 
             'max_high':  sqlal.types.Float(precision=5, asdecimal=True),
             'min_low': sqlal.types.Float(precision=5, asdecimal=True),
             'mean_diff_high_low': sqlal.types.Float(precision=5, asdecimal=True)}
```

```python
df_gold.to_sql("ibm_prices_gold_lauro", engine, if_exists="replace", index=False, dtype=dict_dtype_gold)
```
